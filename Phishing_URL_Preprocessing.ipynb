# Phishing URL Dataset Cleaning and Preprocessing

```
import pandas as pd
import numpy as np
from google.colab import files
uploaded = files.upload()

df = pd.read_csv(list(uploaded.keys())[0])
df
```

```
# checking the data size
print(df.shape)
```

```
# checking the data info
df.info()
```

```
# Correct misspelled column names
df = df.rename(columns={
    'having_IPhaving_IP_Address': 'having_IP_Address',
    'URLURL_Length': 'URL_Length'
})
df
```

```
# checking the null values
df.isnull().sum()
```

```
# Remove unnecessary columns
df = df.drop(columns=['index'])
df
```

```
# Remove duplicates to prevent data leakage from training to test, and reduce overhead on the model
print("Duplicates after removing index:", df.duplicated().sum())
df = df.drop_duplicates()
```

```
# Check the data size after removing Duplicates
print(df.shape)
```

```
# Encode the target column (y) from text to numerical values for ML/DL models
df.loc[:, 'Result'] = df['Result'].map({
    'Phishing': 1,
    'Legitimate': 0
})
df['Result'].value_counts()
```

```
# split the dataset into training and testing sets

from sklearn.model_selection import train_test_split

X = df.drop(columns=['Result'])
y = df['Result']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=0,
    stratify=y
)

print(X_train.shape)
print(X_test.shape)
```

```
# Balance classes using undersampling to prevent model bias

from sklearn.utils import resample

train_df = X_train.copy()
train_df['Result'] = y_train.values

class_0 = train_df[train_df['Result'] == 0]
class_1 = train_df[train_df['Result'] == 1]

class_0 = resample(class_0, replace=False, n_samples=len(class_1), random_state=42)

train_df = pd.concat([class_0, class_1]).sample(frac=1, random_state=42).reset_index(drop=True)

X_train = train_df.drop(columns=['Result'])
y_train = train_df['Result']

print(y_train.value_counts())
```

```
print(y_train.value_counts())
```

```
# Feature scaling using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```
print("Preprocessing completed successfully.")
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```
y_train = y_train.astype(int)
y_test = y_test.astype(int)
```

Baseline Model: Logistic Regression

Logistic Regression was used as the baseline model to provide a reference point for comparing the performance of improved machine learning and deep learning models.

```
# Baseline Model: Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

# Create baseline model
baseline_model = LogisticRegression(max_iter=1000, random_state=42)

# Train model
baseline_model.fit(X_train, y_train)

# Predict
y_pred = baseline_model.predict(X_test)

# Evaluation
print("- Logistic Regression Baseline Results -")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1-Score:", f1_score(y_test, y_pred))

print("\n - Confusion Matrix - ")
print(confusion_matrix(y_test, y_pred))

print("\n - Classification Report -")
print(classification_report(y_test, y_pred))
```

# Results Analysis

The Logistic Regression baseline model achieved good phishing URL classification performance.

The model achieved:

 • Accuracy: 92.05%

 • Precision: 91.71%

 • Recall: 91.87%

 • F1-Score: 91.79%

The results indicate that Logistic Regression provided balanced classification performance and served as an effective baseline model for comparison with the improved models.



# Confusion Matrix Analysis:

The confusion matrix showed that the Logistic Regression model correctly classified 557 legitimate URLs and 520 phishing URLs. However, 47 legitimate URLs were misclassified as phishing, while 46 phishing URLs were classified as legitimate. These results indicate that the model achieved balanced classification performance with a relatively low number of classification errors.

```
# Improved Model (Machine Learning)


from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score
from sklearn.metrics import recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt


# Create the Random Forest model

rf_model = RandomForestClassifier(
n_estimators=101,
max_depth=10,
random_state=42
)

# Train the model using the training dataset

rf_model.fit(X_train, y_train)

# Generate predictions using the test dataset

y_pred_rf = rf_model.predict(X_test)
```

### Random Forest Evaluation

The Random Forest model was evaluated using multiple classification metrics, including Accuracy, Precision, Recall, F1-Score, Confusion Matrix, and Classification Report.

These metrics were used to measure the model’s ability to correctly distinguish phishing URLs from legitimate URLs and compare its performance with the Logistic Regression baseline model.

```
# EVALUATION

print("-Random Forest Results-")

print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Precision:", precision_score(y_test, y_pred_rf))
print("Recall:", recall_score(y_test, y_pred_rf))
print("F1-Score:", f1_score(y_test, y_pred_rf))

print("AUC Score:", roc_auc_score(
    y_test,
    rf_model.predict_proba(X_test)[:,1]
))

print("\n-Confusion Matrix-")
print(confusion_matrix(y_test, y_pred_rf))





print("\n-Classification Report-")
print(classification_report(y_test, y_pred_rf))
```

```
# Cross Validation for Random Forest

from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(
    rf_model,
    X_train,
    y_train,
    cv=5,
    scoring='accuracy'
)

print("Cross Validation Scores:", cv_scores)
print("Average CV Accuracy:", cv_scores.mean())
```

## Part 4: Deep Learning Model

### Neural Network Architecture and Design Justification
The ANN model was designed for binary classification of phishing and legitimate URLs using multiple hidden layers and a Sigmoid activation function in the output layer. Different architectures and hyperparameter settings were tested, including variations in the number of neurons, dropout rates, batch sizes, and training epochs, in order to improve model performance and reduce overfitting.

The experiments aimed to identify the most effective ANN configuration for phishing URL detection while maintaining stable training and good generalization on unseen data.

### Experiment 1 – Initial ANN Model

```
import random
import numpy as np
import tensorflow as tf

random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)
```

```
# Deep Learning Model (ANN)

import tensorflow as tf


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report

# Convert labels to integer
y_train = y_train.astype(int)
y_test = y_test.astype(int)

# Build ANN Model
model = Sequential()

model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))

# Compile Model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Train Model
history = model.fit(
    X_train,
    y_train,
    epochs=15,
    batch_size=32,
    validation_split=0.2
)

# Predictions
y_pred_prob = model.predict(X_test)
y_pred_ann1 = (y_pred_prob > 0.5).astype(int)
```

### Model Evaluation

```
# Evaluation
print("\n-ANN Experiment 1 Results-")

y_pred_ann1 = y_pred_ann1.ravel()

ann1_accuracy = accuracy_score(y_test, y_pred_ann1)
ann1_precision = precision_score(y_test, y_pred_ann1)
ann1_recall = recall_score(y_test, y_pred_ann1)
ann1_f1 = f1_score(y_test, y_pred_ann1)

print("Accuracy:", ann1_accuracy)
print("Precision:", ann1_precision)
print("Recall:", ann1_recall)
print("F1-Score:", ann1_f1)

print("\n-Confusion Matrix-")
print(confusion_matrix(y_test, y_pred_ann1))

print("\n-Classification Report-")
print(classification_report(y_test, y_pred_ann1))
```

### Experiment 1: Result Analysis
The initial ANN model achieved approximately 93% accuracy. The model showed balanced Precision and Recall scores, indicating acceptable phishing URL detection performance. However, there was still room for improvement in overall stability and classification performance.

## Experiment 2: Tuned ANN Model

```
# Experiment 2: Tuned ANN Model

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

# Build Tuned ANN Model
ann_tuned = Sequential()

ann_tuned.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
ann_tuned.add(BatchNormalization())
ann_tuned.add(Dropout(0.2))

ann_tuned.add(Dense(64, activation='relu'))
ann_tuned.add(BatchNormalization())
ann_tuned.add(Dropout(0.2))

ann_tuned.add(Dense(32, activation='relu'))
ann_tuned.add(Dropout(0.1))

ann_tuned.add(Dense(1, activation='sigmoid'))

# Compile Model
ann_tuned.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Train Model
history_tuned = ann_tuned.fit(
    X_train,
    y_train,
    epochs=30,
    batch_size=32,
    validation_split=0.2
)

# Predictions
y_pred_prob_tuned = ann_tuned.predict(X_test)
y_pred_ann2 = (y_pred_prob_tuned > 0.5).astype(int)

```

### Model Evaluation

```
# Evaluation
print("=== ANN Experiment 2 Results ===")

y_pred_ann2 = y_pred_ann2.ravel()

ann2_accuracy = accuracy_score(y_test, y_pred_ann2)
ann2_precision = precision_score(y_test, y_pred_ann2)
ann2_recall = recall_score(y_test, y_pred_ann2)
ann2_f1 = f1_score(y_test, y_pred_ann2)

print("Accuracy:", ann2_accuracy)
print("Precision:", ann2_precision)
print("Recall:", ann2_recall)
print("F1-Score:", ann2_f1)

print("\n=== Confusion Matrix ===")
print(confusion_matrix(y_test, y_pred_ann2))

print("\n=== Classification Report ===")
print(classification_report(y_test, y_pred_ann2))
```

### Experiment 2: Result Analysis
In this experiment, the ANN hyperparameters were adjusted to improve the model performance. After tuning the network structure and training settings, the model achieved approximately 94% accuracy with improved Precision, Recall, and F1-Score values. The results demonstrated better overall stability and more balanced classification performance compared to the first experiment.

### Experiment 3: Final Optimized ANN Model





```
# Experiment 3: Wider ANN Model
# Build ANN model
ann_exp3 = Sequential()

# First hidden layer
ann_exp3.add(Dense(256, activation='relu', input_shape=(X_train.shape[1],)))

# Dropout layer to reduce overfitting
ann_exp3.add(Dropout(0.15))

# Second hidden layer
ann_exp3.add(Dense(128, activation='relu'))

# Dropout layer
ann_exp3.add(Dropout(0.15))

# Third hidden layer
ann_exp3.add(Dense(64, activation='relu'))

# Additional dropout layer
ann_exp3.add(Dropout(0.1))

# Output layer for binary classification
ann_exp3.add(Dense(1, activation='sigmoid'))

# Compile model
ann_exp3.compile(
    optimizer=Adam(learning_rate=0.0005),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Train model
history_exp3 = ann_exp3.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=16,
    validation_split=0.2
)

y_pred_prob_exp3 = ann_exp3.predict(X_test)
y_pred_Fann = (y_pred_prob_exp3 > 0.5).astype(int)
```

### Model Evaluation

```
# Model evaluation
print("=== Final ANN Results ===")
print("Accuracy:", accuracy_score(y_test, y_pred_Fann))
print("Precision:", precision_score(y_test, y_pred_Fann))
print("Recall:", recall_score(y_test, y_pred_Fann))
print("F1-Score:", f1_score(y_test, y_pred_Fann))

print("\n=== Confusion Matrix ===")
print(confusion_matrix(y_test, y_pred_Fann))

print("\n=== Classification Report ===")
print(classification_report(y_test, y_pred_Fann))
```

### Experiment 3 - Multiple Runs Average

```
# =========================
# Final ANN Results
# =========================

Fann_accuracy = accuracy_score(y_test, y_pred_Fann)
Fann_precision = precision_score(y_test, y_pred_Fann)
Fann_recall = recall_score(y_test, y_pred_Fann)
Fann_f1 = f1_score(y_test, y_pred_Fann)

print("=== Final ANN Results ===")
print("Accuracy:", Fann_accuracy)
print("Precision:", Fann_precision)
print("Recall:", Fann_recall)
print("F1-Score:", Fann_f1)


# =========================
# Average Results from ANN Experiments
# =========================

accuracies = [ann1_accuracy, ann2_accuracy, Fann_accuracy]
precisions = [ann1_precision, ann2_precision, Fann_precision]
recalls = [ann1_recall, ann2_recall, Fann_recall]
f1_scores = [ann1_f1, ann2_f1, Fann_f1]

# Calculate averages
avg_accuracy = np.mean(accuracies)
avg_precision = np.mean(precisions)
avg_recall = np.mean(recalls)
avg_f1 = np.mean(f1_scores)

# Print final average results
print("\n=== Average Results for ANN Experiments ===")
print(f"Average Accuracy : {avg_accuracy:.2f}")
print(f"Average Precision: {avg_precision:.2f}")
print(f"Average Recall   : {avg_recall:.2f}")
print(f"Average F1-Score : {avg_f1:.2f}")
```
